### Problem Statement 2: Search Query Keyword Analysis

```python
string = "Buy mobile phone buy phone online"
words = string.lower().split()
count = {}
for w in words:
    count[w] = count.get(w, 0) + 1
repeated = {k: v for k, v in count.items() if v > 1}
print(repeated)
```

### Problem Statement 3: Sensor Data Validation

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
valid = []
for i, reading in enumerate(sensor_readings):
    if reading % 2 == 0:   # check even
        valid.append((i, reading))
print(valid)
```

### Problem Statement 4: Email Domain Usage Analysis

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]
domain_count = {}
for email in emails:
    domain = email.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1
total = len(emails)

for domain, count in domain_count.items():
    percent = (count / total) * 100
    print(f"{domain}:{percent:.0f}%")
```

### Problem Statement 5: Sales Spike Detection

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]
avg = sum(sales) / len(sales)
threshold = avg * 1.30
for i, value in enumerate(sales, start=1):
    if value > threshold:
        print(f"Day {i}: {value}")
```
