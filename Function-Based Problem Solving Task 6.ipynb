### Problem Statement 1:Smart Parking Lot Management System

```python
def smart_parking_system(capacity, vehicle_logs):
    current = 0
    for log in vehicle_logs:
        if log == "IN":
            current += 1
        elif log == "OUT":
            current -= 1
    print("Currently Parked Vehicles:", current)
    if current > capacity:
        print("Parking Status: FULL / OVER CAPACITY")
    else:
        print("Parking Status: Available")
capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
smart_parking_system(capacity, vehicle_logs)
```

###  Problem Statement 4:Cloud Server Load Classification System

```python
CPU_readings = [45, 60, 70, 85, 90]
def server_load(CPU_readings):
    total_reading = 0
    for reading in CPU_readings:
        total_reading += reading
    average_reading = total_reading / len(CPU_readings) 
    print(f"Average CPU Load: {int(average_reading)}%") 
    if average_reading < 50:
        print("Server Status: Normal")  
    elif 50 <= average_reading <= 80:
        print("Server Status: Warning") 
    else:
        print("Server Status: Critical")
server_load(CPU_readings)
```

###  Problem Statement 5:Smart Classroom Resource Usage Monitor

```python
Resource_Usage={"Projector": 6,
                        "AC": 9,
                        "Lights": 4
                       }
def expense_tracker(Resource_usage):
    highest_usage=max(Resource_Usage.values())
    highly_use_equipment=[resource for resource , hr in Resource_Usage.items() if hr==highest_usage]
    resource=" , ".join(highly_use_equipment)
    print(f"Overused Resources: {resource}")
    print("Energy Alert : Yes")
expense_tracker(Resource_Usage)
```

### Problem Statement 6: Online Event Registration Capacity Controller

```python
def event_registration(capacity, registrations):
    confirmed = 0
    waitlisted = 0
    for i in range(registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1
    print(f"Confirmed Registrations: {confirmed}")
    print(f"Waitlisted Users: {waitlisted}")
    if confirmed == capacity:
        print("Registration Status: Closed")
    else:
        print("Registration Status: Open")
event_capacity = 100
total_registrations = 105
event_registration(event_capacity, total_registrations)
```
