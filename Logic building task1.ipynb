### Pass / Fail Analyzer

```python
# From the given list of marks analyze the students pass or fail and display their count 
marks = [45, 78, 90, 33, 60]
Pass=0
Fail=0
for m in marks:
    if m>=50:
       Pass+=1
    else:
        Fail+=1
print("Pass:",Pass)
print("Fail:",Fail)
```

```python
# clean the data
names = [" Alice ", "bob", " CHARLIE "]
Cleaned_list=[name.strip().lower() for name in names]
print(Cleaned_list)
```

```python
# Message Length Analyzer
messages = ["Hi", "Welcome to the platform", "OK"]
Flag=False
Length=[len(message) for message in messages]
for i in Length:
    if i>10:
        Flag=True
        print(Length , Flag)
```

```python
# Error Message Detector
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
count=0
for log in logs:
    if log=="ERROR":
        count+=1
print(count)
```
