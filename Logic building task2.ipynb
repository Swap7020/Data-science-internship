### Unique words in a sentence

```python
sentence='python is easy and python is powerful'
sent=sentence.split()
word=set()
for i in sent:
    word.add(i)
print(word)
print(len(word))
```

### Highest Salary from Employee Data

```python
Employee={'Ravi':75000,'Anita':68000,'Kiran':72000,'Swapnil':88000}
name = max(Employee, key=Employee.get)
print(name, "-", Employee[name])
```

### Find Maximum and Minimum Values

```python
numbers=[45,50,60,70,30,10,20,110]
maximum=numbers[0]
minimum=numbers[0]
for num in numbers:
    if num>maximum:
        maximum=num
    if num<minimum:
        minimum=num
print('maximum:',maximum)
print('minimum:',minimum)
```

### Count Products Above a Price Threshold

```python
prices = [3450, 100, 899, 150, 3300,459,2500,1200,999,1100,12500]
Products_above_1000=0
for price in prices:
    if price>1000:
        Products_above_1000+=1
print('Products_above_1000:',Products_above_1000)     
```

### Calculate Attendance Percentage

```python
Attendence=['P','A','P','A','P','A','P','A','A','P']
count=0
for A in Attendence:
    if A=='P':
        count+=1
print((count/len(Attendence )*100)) 
```

### Count Character Frequency

```python
sentence='asxsksiiaknasaajawdoewdiwoxnxjxwxowjwiiqsmzmoasixoxa'
freq = {}
for ch in sentence:
    if ch in freq:
        freq[ch] += 1
    else:
        freq[ch] = 1
print(freq)
```
