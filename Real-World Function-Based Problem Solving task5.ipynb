### Problem 1: Social Media – Post Engagement Analyzer

```python
def analyze_engagement(likes_list):
    total_likes = 0
    for likes in likes_list:
        total_likes += likes
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
    print("Total Likes:", total_likes)
    print("Post Status:", status)
likes = [200, 150, 300, 250, 150]
analyze_engagement(likes)
```

### Problem 3: Agriculture – Rainfall Adequacy Checker

```python
def analyze_rainfall(rainfall_list, required_level):
    total = 0
    for rain in rainfall_list:
        total += rain
    average = total / len(rainfall_list)
    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    print("Average Rainfall:", int(average))
    print("Rainfall Status:", status)
rainfall_data = [65, 80, 70, 73]
analyze_rainfall(rainfall_data, 70)
```

### Problem 4: Social Media – Duplicate Account Detection

```python
def detect_duplicate_accounts(usernames):
    unique_users = set(usernames)
    if len(unique_users) != len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")
users = ["rahul", "amit", "sneha", "rahul", "rohit"]
detect_duplicate_accounts(users)
```

### Problem 9: E-Learning – Student Attendance Eligibility System

```python
def Eligibility(Attendance):
    Present_count = 0      
    for A in Attendance:
        if A == "P":
            Present_count += 1
    percentage = (Present_count / len(Attendance)) * 100  
    print("Attendance Percentage:", percentage)
    if percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")
Attendance = ["P", "A", "P", "P", "A", "P"]
Eligibility(Attendance)
```

### Problem 10: Smart Electricity Bill Analyzer

```python
def electricity_bill(units):
    bill = 0
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7
    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"
    return bill, usage
units = int(input("Enter electricity units consumed: "))
bill_amount, usage_status = electricity_bill(units)
print("Total Bill: ₹", bill_amount)
print("Usage Level:", usage_status)
```
