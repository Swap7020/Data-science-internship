### Problem 1: Mobile Recharge Validation System

```python
def validate_recharge(amount, plan):
    return amount >= 50 and plan in [199, 299, 399, 599]


while True:
    amt = int(input("Enter amount: "))
    plan = int(input("Enter plan: "))

    if validate_recharge(amt, plan):
        print("Recharge Successful ✅")
        break
    else:
        print("Invalid details — Try again ❌")
        
```

### Problem 2: Inventory Reorder Alert System

```python
mart_stock = {
    "Rice": 100,
    "Wheat": 80,
    "Sugar": 50,
    "Milk": 9,
    "Bread": 30,
    "Oil": 25
}

for stock in mart_stock.values():
    if stock<15:
        print('Reorder Alert')
    else:
        print('Stock Ok')
```

###  Problem 3: Student Result Processing System

```python
def process_result(marks):
    total = 0
    for m in marks:
        total += m

    average = total / len(marks)
    if average >= 50:
        return average, "Pass"
    else:
        return average, "Fail"
student_marks = [60, 45, 70, 55]

avg, result = process_result(student_marks)
print("Result:", result)
```

### Problem 4: Cab Fare Estimator with Retry Option

```python
def calculate_fare(km, peak):
    fare = 50 + (12 * km)
    if peak == "yes":
        fare = fare * 1.25
    return fare


while True:
    km = float(input("Enter distance: "))
    peak = input("Peak hour (yes/no): ")

    if km > 0:
        print("Total Fare =", calculate_fare(km, peak))
        break
    else:
        print("Invalid distance, try again")
```

### Problem 5: Employee Attendance Eligibility Checker

```python
def attendance(sheet):
    present = 0
    for s in sheet:
        if s=='P':
            present += 1

    percentage =  present/len(sheet)*100 
    if percentage >= 75:
        return percentage, "Eligible"
    else:
        return percentage, "Not Eligible"
sheet = ["P","A","A","P","P","P","A","P","A","P","P","P","P","P","P","P"]

per, result = attendance(sheet)
print("Result:", result)
```

### Problem 6: Password Strength Checker

```python
def check_password(password):
    has_digit = False
    has_special = False

    if len(password) < 8:
        return " Weak Password (Too short)"

    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in "@#$":
            has_special = True

    if has_digit and has_special:
        return " Strong Password"
    else:
        return " Weak Password (Missing digit or special character)"
pwd = input("Enter password: ")
print(check_password(pwd))
```
